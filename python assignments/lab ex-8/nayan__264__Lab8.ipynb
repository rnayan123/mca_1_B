Q2. Create the data visualization using the matplotlib.




```python
import matplotlib.pyplot as plt
import numpy as np
```

Visualize the Gender of Passengers using the Bar graph.

```python
gender_counts = df['Sex'].value_counts()

fig = plt.figure(figsize = (5, 5))

plt.bar(gender_counts.index, gender_counts.values,width=0.8)
plt.title('Gender of Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
```

Visualize the Survival Count of Passengers using the Bar graph.

```python
survival_counts = df['Survived'].value_counts()
# print(survival_counts)
# print(survival_counts[0]=="Not Survived")
sns.barplot(x = survival_counts.index,y = survival_counts.values)


plt.title('Survival Count of Passengers')
plt.xlabel('Survival')
plt.ylabel('Count')
plt.show()
```

Visualize the Age of Passengers using the Bar/Histogram graph.


```python
plt.hist(df['Age'].dropna(), bins=20, edgecolor = "black")
plt.title('Age of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Visualize the comparison of Age and Fare of Passengers using the Scatterplot.

```python
plt.scatter(df['Age'], df['Fare'])
plt.title('Comparison of Age and Fare of Passengers')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
