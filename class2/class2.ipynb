```python
import numpy as np
from numpy import linspace
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
from scipy import signal
from mpl_toolkits.basemap import Basemap
%matplotlib widget
data=pd.read_csv('Iris.csv')
print(data.head())
print(data.shape)
print(data.describe())
print(data.info())

```

```python
fig =plt.figure()

# syntax for 3-d projection
ax = plt.axes(projection='3d')

#defining all 3 axis
x = data['PetalWidthCm']
y = data['PetalLengthCm']
z = data['SepalLengthCm']
sortX = x.sort_values()
sortY = y.sort_values()
sortZ = z.sort_values()


#plotting
ax.plot3D(sortX,sortY,sortZ,c='green')
ax.set_title('3d Scatter plot using Iris Dataset')
ax.set_xlabel('PetalWidthCm')
ax.set_ylabel('PetalLengthCm')
ax.set_zlabel('SepalLengthCm')
plt.show()
```

```python
setosa=data[data['Species']=='Iris-setosa']
versicolor=data[data['Species']=='Iris-versicolor']
virginica=data[data['Species']=='Iris-virginica']
```

```python

# mpl.rcParams['font.size'] = 10
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.bar(setosa['Id'],setosa['PetalWidthCm'],zs=1,zdir='x',color='red',alpha=0.8)
ax.bar(versicolor['Id'],setosa['PetalWidthCm'],zs=2,zdir='x',color='green',alpha=0.8)
ax.bar(setosa['Id'],setosa['PetalWidthCm'],zs=3,zdir='x',color='blue',alpha=0.8)
ax.set_xlabel('Species')
ax.set_ylabel('Id')
ax.set_zlabel('PetalWidthCm')
plt.show()
# font size
#fig addsubplot  how many subplots we want 111 indicates ocuppies 1 figure 
```
