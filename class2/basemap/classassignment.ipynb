```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
plt.figure(figsize=(10, 8))

m = Basemap(projection='lcc',
            resolution='l',
            lat_0=12.9716,
            lon_0=77.5946,
            width=500000,
            height=900000)
m.shadedrelief()
bengaluru_lat, bengaluru_lon = 12.9716, 77.5946
x, y = m(bengaluru_lon, bengaluru_lat) 
m.plot(x, y, 'bo', markersize=8, label='Bengaluru' ,color="red")

plt.text(x, y, 'Bengaluru', fontsize=12, color='black', ha='right')

m.drawstates(color='k', linewidth=2)
m.drawrivers(color="blue")


m.drawcoastlines()
m.drawmapboundary()
plt.legend()

plt.title('Bangalore in Karnataka')

plt.show()
```

```python
# Create a figure and axis with specified size
fig, ax = plt.subplots(figsize=(10, 12))

# Define the map boundaries for Karnataka
karnataka_map = Basemap(
    projection='merc',
    llcrnrlat=11.5, urcrnrlat=18.5,
    llcrnrlon=74, urcrnrlon=78.5,
    resolution='l'
)

# Draw coastlines, countries, and states
karnataka_map.drawcoastlines(linewidth=0.7, color='black')
karnataka_map.drawcountries(linewidth=1.2, color='black')
karnataka_map.drawstates(linewidth=0.5, color='gray', linestyle='dashed')
karnataka_map.shadedrelief()

# Draw rivers with a blue color
karnataka_map.drawrivers(color='blue', linewidth=0.5)

# Add a marker for Bengaluru with label
bengaluru_lat, bengaluru_lon = 12.9716, 77.5946
x, y = karnataka_map(bengaluru_lon, bengaluru_lat)
karnataka_map.plot(x, y, 'bo', markersize=8, label='Bengaluru')
plt.text(x, y, 'Bengaluru', fontsize=12, ha='right', va='bottom', color='blue', weight='bold')

# Draw parallels and meridians with labels
karnataka_map.drawparallels(range(0, 20, 2), labels=[1, 0, 0, 0], fontsize=10, color='red')
karnataka_map.drawmeridians(range(72, 80, 2), labels=[0, 0, 0, 1], fontsize=10, color='red')

# Add a title
plt.title('Map of Karnataka', fontsize=16, weight='bold')

# Add legend
plt.legend(loc='upper left')

# Display the map
plt.show()
```
