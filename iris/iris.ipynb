```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the CSV file into a DataFrame
data = pd.read_csv('iris.csv')

data.dropna(subset=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm'], inplace=True)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['SepalLengthCm'], data['SepalWidthCm'], data['PetalLengthCm'],cmap="plasma" ,s=50)

ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalLengthCm')
ax.set_title('3D Scatter Plot: SepalLengthCm, SepalWidthCm, PetalLengthCm')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



data = pd.read_csv('iris.csv')

data.dropna(subset=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], inplace=True)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(data['SepalLengthCm'], data['SepalWidthCm'], data['PetalLengthCm'], cmap='viridis', linewidth=0.2, edgecolor='k', antialiased=True)

ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel('PetalLengthCm')
ax.set_title('3D Area Plot: SepalLengthCm, SepalWidthCm, PetalLengthCm')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

Type= list(data.groupby('Species')['SepalLengthCm'].mean())
Avg=list(data.groupby('Species')['SepalLengthCm'].mean())
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111,projection='3d')
colors=['orange','blue','pink']
ax.bar(Type,Avg,Avg,color=colors,zdir='y',alpha=0.5)#alpha is used for transparency
ax.set_title('3D Bar Plot')
ax.set_xlabel('Species')
ax.set_ylabel('SepalWidthCm')
ax.set_zlabel("AverageSepalLengthCm")
plt.show()
```
