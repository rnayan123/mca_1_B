```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset from the CSV file
data = 'Uscereal.csv'
df = pd.read_csv(data)

bar_width = 0.4  

plt.figure(figsize=(10, 6))
plt.bar(df['mfr'], df['fibre'], color='skyblue', width=bar_width)
plt.title('Bar Graph: mfr vs. fiber')
plt.xlabel('mfr')
plt.ylabel('fibre')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['sodium'], df['sugars'], color='green', marker='o')
plt.title('Scatter Plot: sodium  vs sugars')
plt.xlabel('sodium')
plt.ylabel('sugars')
plt.grid(axis='y')
plt.show()
```
