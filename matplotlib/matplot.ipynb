```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
data = pd.read_csv('Uscereal.csv')

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='mfr', y='calories', data=data)
plt.title('Average Calories by Manufacturer')
plt.show()
```

```python


data.dropna(subset=['protein', 'fat'], inplace=True)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='protein', y='fat', hue='protein', palette='viridis', data=data)
plt.xlabel('Protein (g)')
plt.ylabel('Fat (g)')
plt.title('Scatter Plot: Protein vs Fat with Different Colors')
plt.legend(title='Protein (g)', loc='upper right')
plt.show()

```

```python



plt.figure(figsize=(10, 6))

sns.histplot(data['calories'], color='blue', alpha=0.7, bins=20, label='Calories')
sns.histplot(data['protein'], color='orange', alpha=0.7, bins=20, label='Protein')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.title('Defined Histogram: Calories and Protein')

plt.show()
```

```python
plt.pie(data['mfr'].value_counts(), labels=data['mfr'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Manufacturers')
plt.show()
```

```python
sns.pairplot(data[['calories', 'protein', 'fat', 'fibre']])
plt.suptitle('Pair Plot of Nutritional Content')
plt.show()
```

```python
plt.scatter(data['protein'], data['fat'], s=data['calories']*5, alpha=0.5)
plt.xlabel('Protein (g)')
plt.ylabel('Fat (g)')
plt.title('Bubble Plot: Protein vs Fat with Calories')
plt.show()
```

```python
sns.boxplot(x='mfr', y='calories', data=data)
plt.title('Box Plot: Calories by Manufacturer')
plt.show()
```

```python
sns.violinplot(x='mfr', y='calories', data=data)
plt.title('Violin Plot: Calories by Manufacturer')
plt.show()
```

```python
plt.stackplot(data['mfr'].unique(),
              [data[data['mfr'] == m]['calories'].mean() for m in data['mfr'].unique()],
              labels=data['mfr'].unique())
plt.xlabel('Manufacturer')
plt.ylabel('Average Calories')
plt.title('Area Plot: Average Calories by Manufacturer')
plt.legend()
plt.show()
```
